**Perform the following operations using Python data sets** 

**a. Create data subsets**

**b. Merge Data**

**c. Sort Data**

**d. Transposing Data**

**e. Shape and reshape Data** 

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("temperatures.csv")

df
```

```python
# ----------------------------------------
# a. Create Data Subsets
# ----------------------------------------

# Select specific columns
subset_cols = df[['JAN', 'FEB']]

# Select rows using index
subset_rows = df.iloc[10:30]

# Conditional subset
subset_cond = df[df['MAR'] > 30]

subset_rows
```

```python
# ----------------------------------------
# b. Merge Data
# ----------------------------------------

# Create sample subsets
df1 = df.iloc[0:20]
df2 = df.iloc[20:40]

# Merge (concatenate)
merged_df = pd.concat([df1, df2])


merged_df
```

```python
# ----------------------------------------
# c. Sort Data
# ----------------------------------------

# Sort by a column (ascending/descending)
sorted_df = df.sort_values(by='ANNUAL', ascending=False)

sorted_df
```

```python
# ----------------------------------------
# d. Transposing Data
# ----------------------------------------

# Transpose dataframe (rows ↔ columns)
transposed_df = df.T

transposed_df
```

```python
# ----------------------------------------
# e. Shape and Reshape Data
# ----------------------------------------

# Get shape (rows, columns)
print(df.shape)

# Convert DataFrame to NumPy array
arr = df.to_numpy()

# Reshape array (example)
reshaped_arr = arr.reshape(-1)
print(reshaped_arr)
# Example: reshape into 2 columns (only if size allows)
# reshaped_arr = arr.reshape(rows, cols)
# Replace 'col1', 'col2' with actual column names from your dataset.
# Reshaping only works if the total number of elements matches.
```
