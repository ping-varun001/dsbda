```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Heart_Disease_Prediction.csv")
df = df.select_dtypes(include=['number'])
df
```

```python
# ----------------------------------------
# a. Data Cleaning
# ----------------------------------------

# Rename columns
df = df.rename(columns={'Cholesterol': 'chol'})

# Drop unwanted columns
df = df.drop(['Exercise angina'], axis=1)

# Remove duplicates
df = df.drop_duplicates()

# Convert data types (example)
df['BP'] = df['BP'].astype(float)


```

```python
# ----------------------------------------
# b. Data Integration
# ----------------------------------------

# Load another dataset
df2 = pd.read_csv("heart.csv")

# Merge datasets (common column required)
merged_df = pd.merge(df, df2, on='chol', how='inner')

# OR concatenate
concat_df = pd.concat([df, df2])
concat_df
```

```python
# ----------------------------------------
# c. Data Transformation
# ----------------------------------------

# Scaling / transformation
df['EKG results'] = df['EKG results'] * 100

# Log transformation
df['Age'] = np.log(df['Age'])

# Create new feature
df['new_col'] = df['Age'] + df['Sex']
```

```python
# ----------------------------------------
# d. Error Correcting
# ----------------------------------------
sns.boxplot(data=df)
plt.title("Box Plot")
plt.show()



```

```python
def remove_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Keep rows where all values are within bounds
    data_clean = data[~((data < lower_bound) | (data > upper_bound)).any(axis=1)]
    
    return data_clean

# Apply outlier removal
df_clean = remove_outliers_iqr(df)

print("Original shape:", df.shape)
print("After removing outliers:", df_clean.shape)
```

```python
# ----------------------------------------
# e. Data Model Building
# ----------------------------------------

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Define features and target
X = df.drop('Thallium', axis=1)
y = df['Thallium']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

#Replace 'col1', 'target_column', 'common_column' with actual column names.
```

```python
results = pd.DataFrame({
    'Actual Value': y_test,
    'Predicted Value': y_pred
})

results[:10]
```
