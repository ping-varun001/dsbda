**Visualize the data using Python libraries matplotlib,seaborn by plotting the graphs**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("your_dataset.csv")

# Select numeric columns
df_num = df.select_dtypes(include=np.number)


# ----------------------------------------
# 1. Line Plot
# ----------------------------------------
plt.plot(df_num.iloc[:10, 0], df_num.iloc[:10, 1])
plt.title("Line Plot")
plt.show()


# ----------------------------------------
# 2. Scatter Plot
# ----------------------------------------
plt.scatter(df_num.iloc[:, 0], df_num.iloc[:, 1])
plt.title("Scatter Plot")
plt.show()


# ----------------------------------------
# 3. Bubble Plot ✅
# ----------------------------------------
plt.scatter(
    df_num.iloc[:, 0],
    df_num.iloc[:, 1],
    s=df_num.iloc[:, 2] * 10,
    c=df_num.iloc[:, 2],
    cmap='viridis',
    alpha=0.6
)
plt.colorbar()
plt.title("Bubble Plot")
plt.show()


# ----------------------------------------
# 4. Bar Plot
# ----------------------------------------
plt.bar(df_num.iloc[:10, 0], df_num.iloc[:10, 1])
plt.title("Bar Plot")
plt.show()


# ----------------------------------------
# 5. Horizontal Bar Plot
# ----------------------------------------
plt.barh(df_num.iloc[:10, 1], df_num.iloc[:10, 0])
plt.title("Horizontal Bar Plot")
plt.show()


# ----------------------------------------
# 6. Histogram
# ----------------------------------------
plt.hist(df_num.iloc[:, 0], bins=20)
plt.title("Histogram")
plt.show()


# ----------------------------------------
# 7. Pie Chart
# ----------------------------------------
df_num.iloc[:, 0].value_counts().head().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()


# ----------------------------------------
# 8. Box Plot
# ----------------------------------------
sns.boxplot(data=df_num)
plt.title("Box Plot")
plt.show()


# ----------------------------------------
# 9. Violin Plot
# ----------------------------------------
sns.violinplot(y=df_num.iloc[:, 0])
plt.title("Violin Plot")
plt.show()


# ----------------------------------------
# 10. Strip Plot
# ----------------------------------------
sns.stripplot(x=df_num.iloc[:, 0], y=df_num.iloc[:, 1])
plt.title("Strip Plot")
plt.show()


# ----------------------------------------
# 11. Swarm Plot
# ----------------------------------------
sns.swarmplot(x=df_num.iloc[:100, 0], y=df_num.iloc[:100, 1])
plt.title("Swarm Plot")
plt.show()


# ----------------------------------------
# 12. KDE Plot
# ----------------------------------------
sns.kdeplot(df_num.iloc[:, 0], fill=True)
plt.title("KDE Plot")
plt.show()


# ----------------------------------------
# 13. Regression Plot
# ----------------------------------------
sns.regplot(x=df_num.iloc[:, 0], y=df_num.iloc[:, 1])
plt.title("Regression Plot")
plt.show()


# ----------------------------------------
# 14. Heatmap
# ----------------------------------------
sns.heatmap(df_num.corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()


# ----------------------------------------
# 15. Pair Plot
# ----------------------------------------
sns.pairplot(df_num)
plt.show()


# ----------------------------------------
# 16. Area Plot
# ----------------------------------------
df_num.iloc[:10, :3].plot.area()
plt.title("Area Plot")
plt.show()


# ----------------------------------------
# 17. Stem Plot
# ----------------------------------------
plt.stem(df_num.iloc[:10, 0])
plt.title("Stem Plot")
plt.show()


# ----------------------------------------
# 18. Step Plot
# ----------------------------------------
plt.step(df_num.iloc[:10, 0], df_num.iloc[:10, 1])
plt.title("Step Plot")
plt.show()


# ----------------------------------------
# 19. Hexbin Plot
# ----------------------------------------
plt.hexbin(df_num.iloc[:, 0], df_num.iloc[:, 1], gridsize=25, cmap='Blues')
plt.colorbar()
plt.title("Hexbin Plot")
plt.show()


# ----------------------------------------
# 20. Error Bar Plot
# ----------------------------------------
x = np.arange(10)
y = df_num.iloc[:10, 0]
error = np.random.rand(10)

plt.errorbar(x, y, yerr=error, fmt='o')
plt.title("Error Bar Plot")
plt.show()


# ----------------------------------------
# 21. Joint Plot
# ----------------------------------------
sns.jointplot(x=df_num.iloc[:, 0], y=df_num.iloc[:, 1], kind='scatter')
plt.show()


# ----------------------------------------
# 22. Boxen Plot
# ----------------------------------------
sns.boxenplot(y=df_num.iloc[:, 0])
plt.title("Boxen Plot")
plt.show()


# ----------------------------------------
# 23. ECDF Plot
# ----------------------------------------
sns.ecdfplot(df_num.iloc[:, 0])
plt.title("ECDF Plot")
plt.show()


# ----------------------------------------
# 24. Rug Plot
# ----------------------------------------
sns.rugplot(df_num.iloc[:, 0])
plt.title("Rug Plot")
plt.show()


# ----------------------------------------
# 25. Seaborn Line Plot
# ----------------------------------------
sns.lineplot(x=df_num.iloc[:10, 0], y=df_num.iloc[:10, 1])
plt.title("Seaborn Line Plot")
plt.show()


# ----------------------------------------
# 26. Seaborn Bar Plot
# ----------------------------------------
sns.barplot(x=df_num.iloc[:10, 0], y=df_num.iloc[:10, 1])
plt.title("Seaborn Bar Plot")
plt.show()


# ----------------------------------------
# 27. Point Plot
# ----------------------------------------
sns.pointplot(x=df_num.iloc[:10, 0], y=df_num.iloc[:10, 1])
plt.title("Point Plot")
plt.show()
```
